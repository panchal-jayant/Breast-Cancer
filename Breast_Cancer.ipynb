***Model***

```python
dataset.head()
```

```python
x = dataset.drop('diagnosis', axis = 1)
y = dataset['diagnosis']
```

Splitting training and testing data

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.30,random_state=100)
```

Scaling the data for normalization

```python
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

Made a function to fit and predict model and visualize it



```python
def all(model):
  model.fit(x_train,y_train)
  pred = model.predict(x_test)
  acc = accuracy_score(y_test, pred)
  print("Confusion Matrix")
  cm = confusion_matrix(y_test,pred)
  sns.heatmap(cm, annot=True, fmt='d', cmap=['red', 'green'])
  plt.title('Confusion matrix')
  plt.xlabel('Predcted label')
  plt.ylabel('True label')
  plt.show()
  print(cm)
  print(classification_report(y_test,pred))
  print('accuracy_score : ' , acc)
```

Checking diffrent models to obtain best accuracy

***Random Forest***

```python
randomforest=RandomForestClassifier(n_estimators = 100, random_state=5)
all(randomforest)
```

***Naive Bayes***

```python
nb = GaussianNB()
all(nb)
```

***KNN***

```python
knn =KNeighborsClassifier(n_neighbors=5)
all(knn)
```

***Decision Tree***

```python
DT = DecisionTreeClassifier(random_state=42)
all(DT)
```

***Logistic Regression***

```python
LR = LogisticRegression(random_state=5)
all(LR)
```
